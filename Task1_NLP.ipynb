```python
#Task 1: Conceptual Understanding

1.Difference between "Love" and "love":
In NLP, case sensitivity matters. "Love" (capitalized) and "love" (lowercase) are treated as different tokens unless text is normalized. This can lead to inconsistencies in analysis, so converting everything to lowercase is a common preprocessing step.

2.What happens if stopwords are not removed:
Stopwords (like "the", "is", "and") can dominate the vocabulary, increase dimensionality, and dilute meaningful patterns. Models may waste resources processing them, which can reduce efficiency and accuracy in tasks like classification.

3.Two real-world scenarios where removing stopwords can be harmful:
1. **Sentiment analysis:** In phrases like *"not happy"*, removing "not" changes the meaning entirely.
2. **Legal/medical documents:** Words such as *"shall not"* or *"do not"* are critical to obligations or diagnoses. Removing them could distort the interpretation.

4.Difference between stemming and lemmatization:
- **Stemming:** A rule-based process that chops off word endings to reduce them to a root form (e.g., "running" → "run", "studies" → "studi"). It is fast but can produce non-dictionary forms.
- **Lemmatization:** A linguistically informed process that uses vocabulary and grammar to return the base dictionary form (e.g., "running" → "run", "studies" → "study"). It is more accurate but computationally heavier.
```

```python
# Task 2: Build Advanced Preprocessing Function
import re

def preprocess_text(text: str) -> list:
    """
    Preprocess raw text into clean tokens.
    Steps:
    - Convert to lowercase
    - Remove URLs and email-like patterns
    - Remove numbers
    - Handle repeated characters (collapse to single)
    - Remove punctuation/emojis
    - Normalize spaces
    - Tokenize
    - Remove very short tokens (<=2), except 'no' and 'not'
    """
    if not text or text.strip() == "":
        return []

    # Lowercase
    text = text.lower()

    # Remove URLs and email-like patterns
    text = re.sub(r'http\S+|www\S+|@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters (e.g., "soooo" -> "so")
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove punctuation/emojis (keep only letters and spaces)
    text = re.sub(r'[^a-z\s]', ' ', text)

    # Normalize spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Tokenize
    tokens = text.split()

    # Remove very short tokens except "no" and "not"
    tokens = [t for t in tokens if len(t) > 2 or t in ["no", "not"]]

    return tokens
```

```python
# Task 3: Stress Testing

sample_inputs = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

for s in sample_inputs:
    tokens = preprocess_text(s)
    print("Original:", s)
    print("Cleaned Tokens:", tokens)
    print("Cleaned Sentence:", " ".join(tokens))
    print("-"*50)
```

```python
# Task 4: Token Analytics

def token_stats(tokens):
    total = len(tokens)
    unique = len(set(tokens))
    avg_len = sum(len(t) for t in tokens)/total if total > 0 else 0
    return total, unique, avg_len

for s in sample_inputs:
    tokens = preprocess_text(s)
    stats = token_stats(tokens)
    print("Sentence:", s)
    print("Total Tokens:", stats[0])
    print("Unique Tokens:", stats[1])
    print("Average Token Length:", round(stats[2], 2))
    print("-"*50)
```

```python
# Task 4: Analysis Questions

1.Which sentence had the most noise?
- "I absolutely looooved this product 😍😍" → contained repeated characters and emojis.
- "Win $$$ now!!! Limited offer!!!" → contained symbols and repeated punctuation.
Both had heavy noise before cleaning.

2.Which sentence retained the most meaningful tokens after cleaning?
- "I am not happy with this" → retained important sentiment-bearing tokens ("not", "happy").
- This shows why careful stopword handling is critical.
```

```python
# Task 5: Frequency Analysis
from collections import Counter

# Collect all tokens from the sample inputs
all_tokens = []
for s in sample_inputs:
    all_tokens.extend(preprocess_text(s))

# Count frequencies
counter = Counter(all_tokens)

print("Top 10 most frequent words:", counter.most_common(10))
print("Top 5 least frequent words:", counter.most_common()[-5:])
```

```python
### Task 5: Frequency Analysis

**Top 10 most frequent words:**
These are the words that appear most often across all cleaned sentences. They usually represent common sentiment or repeated emphasis (e.g., "now", "best", "offer").

**Top 5 least frequent words:**
These are words that appear only once or very rarely. They often capture unique context from specific sentences (e.g., "course", "service").

**Interpretation:**
- Frequent words highlight recurring themes in the dataset (like urgency in marketing phrases).
- Rare words capture sentence‑specific meaning and can be important for nuanced analysis.
```

```python
# Task 6: Build Full Pipeline

def full_pipeline(text_list):
    """
    Full NLP preprocessing pipeline.
    Input: list of raw text strings
    Output: dictionary with tokens and cleaned sentences
    """
    results = {"tokens": [], "clean_sentences": []}
    for text in text_list:
        tokens = preprocess_text(text)
        results["tokens"].append(tokens)
        results["clean_sentences"].append(" ".join(tokens))
    return results

# Run pipeline on sample inputs
pipeline_output = full_pipeline(sample_inputs)
print("Pipeline Output:")
print(pipeline_output)
```

```python
### Task 6: Full Pipeline

The full pipeline processes each sentence through the preprocessing function and returns:
- **tokens**: list of cleaned tokens for each sentence
- **clean_sentences**: reconstructed cleaned sentences

This modular design makes the pipeline reusable for larger datasets and ensures consistent preprocessing across all inputs.
```
