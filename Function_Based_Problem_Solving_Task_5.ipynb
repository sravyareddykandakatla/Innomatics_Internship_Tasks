```python
# 1.Social Media – Post Engagement Analyzer
def analyze_post_engagement(likes_list):
    # Calculate total likes using a loop
    total_likes = 0
    for like in likes_list:
        total_likes += like
    # Determine post status based on total likes
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"
    # Display results
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")
# Example usage
likes = [200, 300, 250, 300]  # total = 1050
analyze_post_engagement(likes)
```

```python
#2.Healthcare – Medicine Stock Alert System
def check_medicine_stock(stock):
    # Conditional logic to check stock levels
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
    print(f"Medicine Stock: {stock}")
    print(f"Status: {status}")
# Example usage
check_medicine_stock(6)
```

```python
#3.Agriculture – Rainfall Adequacy Checker
def rainfall_checker(rainfall_data, required_level):
    # Calculate total rainfall using a loop
    total_rainfall = 0
    for value in rainfall_data:
        total_rainfall += value
    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)
    # Determine adequacy based on required level
    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print(f"Average Rainfall: {int(average_rainfall)}")
    print(f"Rainfall Status: {status}")
# Example usage
rainfall_data = [60, 80, 70, 78]   # seasonal rainfall values
required_level = 70                # required adequacy level
rainfall_checker(rainfall_data, required_level)
```

```python
# 4.Social Media – Duplicate Account Detection
def detect_duplicates(usernames):
    # Convert list to set to check uniqueness
    unique_usernames = set(usernames)
    # If lengths differ, duplicates exist
    if len(unique_usernames) < len(usernames):
        status = "Duplicate Accounts Found: Yes"
    else:
        status = "Duplicate Accounts Found: No"
    print(status)
# Example usage
usernames = ["alice", "bob", "alice", "charlie"]
detect_duplicates(usernames)
```

```python
# 5.Healthcare – Appointment Eligibility Checker
def check_appointment_eligibility(age):
    # Condition check based on age
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")
# Example usage
check_appointment_eligibility(21)
```

```python
# 6.Agriculture – Premium Crop Price Filter
def premium_crop_filter(crop_prices, threshold=2000):
    # Use list filtering logic to keep only premium crops
    premium_crops = [price for price in crop_prices if price > threshold]
    print(f"Premium Crops: {premium_crops}")
# Example usage
crop_prices = [1800, 2500, 1900, 3200, 1500]
premium_crop_filter(crop_prices)
```

```python
# 7.System Monitoring – Application Health Checker
def application_health_checker(error_count):
    # Conditional checks for system status
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")
# Example usage
application_health_checker(7)
```

```python
# 8.Banking – Daily Transaction Limit Checker
def transaction_limit_checker(amount, daily_limit=50000):
    # Conditional logic to check transaction status
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")
# Example usage
transaction_limit_checker(60000)
```

```python
# 9.E-Learning – Student Attendance Eligibility System
def attendance_checker(attended_classes, total_classes):
    # Calculate attendance percentage using loop
    total_attended = 0
    for cls in attended_classes:
        total_attended += cls
    # Compute percentage
    attendance_percentage = (total_attended / total_classes) * 100
    # Check eligibility
    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {status}")
# Example usage
# Suppose student attended 40 out of 50 classes
attended_classes = [1]*40
total_classes = 50
attendance_checker(attended_classes, total_classes)
```

```python
# 10.Smart Electricity Bill Analyzer
def electricity_bill(units):
    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
    # Classify usage level based on bill amount
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    # Return results
    return bill, usage_status
# Example usage
units_consumed = 250
bill, status = electricity_bill(units_consumed)
print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{bill}")
print(f"Usage Status: {status}")
```
