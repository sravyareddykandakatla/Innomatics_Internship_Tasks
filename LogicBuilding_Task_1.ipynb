```python
# 1.User Login Check
# Given credentials
username = "admin"
password = "1234"
# Input from user
user_input = input("Enter username: ")
pass_input = input("Enter password: ")
# Authentication check
if user_input == username and pass_input == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2.Pass / Fail Analyzer
# Given marks
marks = [45, 78, 90, 33, 60]
# Initialize counters
pass_count = 0
fail_count = 0
# Analyze results
for m in marks:
    if m >= 50:
        pass_count += 1
    else:
        fail_count += 1
# Print final result
print("Total Students:", len(marks))
print("Passed Students:", pass_count)
print("Failed Students:", fail_count)
```

```python
#3.Simple Data Cleaner
# Given names
names = [" Alice ", "bob", " CHARLIE "]
# Clean and standardize
cleaned_names = [name.strip().lower() for name in names]
# Print the cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
#4.Message Length Analyzer
# Given messages
messages = ["Hi", "Welcome to the platform", "OK"]
# Analyze each message
for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}", end=" ")
    if length > 10:
        print("-Flagged (Too Long)")
    else:
        print("-OK")
```

```python
#5.Error Message Detector
# Given logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
# Count the number of "ERROR" entries
error_count = logs.count("ERROR")
# Print the total error count
print("Total Error Messages:", error_count)
```
